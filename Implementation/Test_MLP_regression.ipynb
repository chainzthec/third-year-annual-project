# Perceptron Multi Couches - Regre

```python
import CLibrary as CLib
import matplotlib.lines as mlines
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D
```

```python
def get(i, l):
    return [z[i] for z in l]
```

## Single Class

### Linear Simple 2D :
- <font color='green'>Linear Model : OK</font>
- <font color='green'>MLP (1, 1) : OK</font>

```python
X = np.array([[1],[2]])
Y = np.array([2,3])

XF = X.flatten().tolist()
YF = Y.flatten().tolist()

N = [1, 1]  
mlpReg_1 = CLib.init(N)

sampleCount = 2

CLib.fit_classification(
    mlpReg_1,
    XF, 
    YF,
    sampleCount,
    10000,
    0.01 
)

for x, y in zip(X, Y):
    plt.scatter(
        x,
        y,
        color='red'
    )

for i in range(0, 300):
    i /= 100
    value = CLib.predict(mlpReg_1, [i], N);
    #print(str(i) + " - " + str(value))
    plt.scatter(i, value, marker=',', color='blue', s=1)

plt.show()
plt.clf()
```

### Non Linear Simple 2D :
- <font color='green'>Linear Model : OK</font>
- <font color='green'>MLP (1, ?, 1) : OK</font>

```python
X = np.array([ [1], [2], [3] ]) 
Y = np.array([ 2, 3, 2.5 ])

XF = X.flatten().tolist()
YF = Y.flatten().tolist()

N = [1, 4, 1]  
mlpReg_2 = CLib.init(N)

sampleCount = 3

CLib.fit_classification(
    mlpReg_2,
    XF, 
    YF,
    sampleCount,
    10000,
    0.001 
)

for x, y in zip(X, Y):
    plt.scatter(
        x,
        y,
        color='red'
    )

for i in range(0, 300):
    i /= 100
    plt.scatter(i, CLib.predict(mlpReg_2, [i], N), marker=',', color='blue', s=1)

plt.show()
plt.clf()
```

### Linear Simple 3D :
- <font color='green'>Linear Model : OK</font>
- <font color='green'>MLP (2, 1) : OK</font>

```python
X = np.array([ [1, 1], [2, 2], [3, 3] ]) 
Y = np.array([ 1, 2, 3 ])

XF = X.flatten().tolist()
YF = Y.flatten().tolist()

N = [2, 1]  
mlpReg_3 = CLib.init(N)

sampleCount = 3

CLib.fit_classification(
    mlpReg_3,
    XF, 
    YF,
    sampleCount,
    10000,
    0.001 
)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0], X[:,1], Y, color="red")

XX, YY, ZZ = [], [], []

for i in range(0, 1000, 10):
    for j in range(0, 1000, 10):
        i /= 100
        j /= 100
        XX.append(i)
        YY.append(j)
        zz = CLib.predict(mlpReg_3, [i, j], N)
        ZZ.append(zz[0])

ax.plot_trisurf(XX, YY, ZZ, lw=0, color="grey", alpha=0.5)
plt.show()
```
